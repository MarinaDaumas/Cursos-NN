## Bias and Variance

    High error in the trainig set means a high bias (underfitting). Can be improved by doing more iterations, getting a better fitting architecture or using bigger NN.

    High error in the dev set means high variance (overffiting). Can be improved by adding more data, changing architecture or use regularization.

    Worst case with high bias and high variance. There may be a bias variance trade of, improving one worsens the other, but it doesn't always happen. With big ennough nn it's almost always possible to "treat" each problem separately.

## Regularization methods

### L2 regularization:
   
   Improve the cost function, make it 
    
$$J(w, b) = \frac{1}{m}\sum_{i=0}^m *L(\hat{y_i}, y_i) + \frac{\lambda}{2m}||w||^2_2$$

$$||w||^2_2 = w^t*w$$

To implement it in a nn: Frobinie norm

$||w$<sup>[$l$]</sup>$||^2_2 = \sum_{i=0}^n$<sup>$[l]$</sup>$\sum_{j=0}^n$<sup>[$l-1$]</sup>

$J(w, b) = \frac{1}{m}\sum_{i=0}^m *L(\hat{y_i}, y_i) + \frac{\lambda}{2m}\sum_{l=0}^L||w$<sup>[$l$]</sup>$||^2_2$

$dw^l= $(regular back prop)$+\frac{\lambda}{m}*w$<sup>[$l$]</sup>

### Drop out regularization
    
    Might be very usefull for computer vision. Almost always use dropout.
    
    Some of the neurons will randomly be set to 0 (A1, A2, etc)
    
    Larger layers can have a lower keep probability to reduce overfitting.
     
- inverted drop out

exemple with layer $l = 3$; keeping prob $= 0.8$; $d3 = $ drop out vector

It's important to divide $a$ by keep_prob in order to keep the scale 

```python
import numpy as np

keep_prob = 0.8

a3 = np.random.rand(6, 3)
d3 = np.random.rand(a3.shape[0], a3.shape[1]) < keep_prob

a3 =  np.multiply(a3, d3)
a3 /= keep_prob

print(a3)
```

### Normalize inputs

    Normalizing inputs makes the cost function have a rounder shape, making it easiear to find the minimun of the function. We can then take larger steps at each the gradient descent and have a faster training.

- subtract the mean

$\mu = \frac{1}{m}\sum_{i=1}^mx^i$

$x := x - \mu$ 

- normaliza variances

$\sigma = \frac{1}{m}\sum_{i=1}^mx^i**2$

$x /= \sigma^2$
