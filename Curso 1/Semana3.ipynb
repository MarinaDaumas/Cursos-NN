## Neural networks representation
$$z = w_1*x_1+w_2*x_2 + b  -->  \hat{y} = a = \sigma(y)  -->  L(a, y)$$



This is a 3 layers NN. (the input is not counted)

Layers can be input, hidden or output. Untill now the NN only had input and output.<br/>
A NN with $n$ layers will have each layear represented by $a^[0], a^[1], a^[n]$<br/>


### Computing NN's output
While in one layer NN we had one logistic regression, now each neuron of the next layer will perform a logistic regression (as if they were the output layer). Each of these computations results in a $a_i^l$<br/>
l : number of the layer<br/> 
i : number of the node(neuron) in the layer

so: $$z_2^1 = w_2^1.T * x + b_2^1$$ <br/>
$$a_2^1 = \sigma(z_2^1)$$

To calculate all of the a's efficiently we vectorize everything. $w_i$.T was a 1x$n$ vector that we will stack with other w_i to make W $k$x$n$ matrix, where <br/>
n : number of inputs (nodes) from layer $l-1$<br/>
k : number of nodes in layer $l$

For exemple, to calculate $a^1$ from the image 
$$z^1 =  W^1 * a^0 + B^1$$
$$a^1 = \sigma(z^1)$$

$$z^2 = W^2 * a^1 + B^2$$
$$a^2 = \sigma(z^2) = \hat{y}$$ 

$W^1$ : 6x4 matrix<br/> 
$a^0$ : 4x1 vector<br/>
$z^1$ : 6x1 vector<br/>


Finally $a^2$ is an 1x1 vector












### Vectorizing across multiple exemples
Now we stack together horizontally the computation the same layer of all the trainnig exemples.

$A^[l] = [a^l0, a^l1, a^l2, ..., a^lm ]$ : matrix kxm

$$Z^1 =  W^1 * A^0 + B^1$$
$$A^1 = \sigma(Z^1)$$

$$Z^2 =  W^2 * A^1 + B^2$$
$$A^2 = \sigma(Z^2) = \hat{Y}$$


### Activation functions
Untill now always used sigmoid ($\sigma$) function. It's usually not a very good option. Only when dealing wth the output layer of a binary classification NN.

Better options are RELU, Leaky RELU and Tangh(almost sigmoid, but from -1 to 1).

It's important to have a non linear activation function is important bacause computing linear functions is useless in most cases. The compositions of linear function is another linear function.
