# Semana 3: Object detection

### Classification with localization

Classifies the object in the picture and makes a bounding box arround it. The NN then outputs the class label and the box specifics, $b_x$, $b_y$, $b_h$, $b_w$.



In the case of a classification of 3 classes c, and oly one of them per image

$$y = [p_c, b_x, b_y, b_w, b_h, c_1, c_2, c_3]$$
Where $p_c$ tells if there's an object in the image. If $p_c = 0$ then the other parameters of $y$ are all going to be don't cares. Else all parameters should be compared with the correct answer.


### Landmark detection

For each landmark, there is a pair $l_x, l_y$ in the output for it's localization in the image.



### Sliding window detection
In this algorithm the convnet that detects the object is runs over a window that has its position slightly changed going over all the image, and then many times with diferent window sizes. Runnig a sliding window detection is very expansive and not very precise on drawing the bounding box.

### Fully conected to convolutional layers

Instead of using fully conected layers you may see them as result of a convolution by a [h, w, c] filter where c is equal to the number of nodes in the equivalent fc layer and h, w are the height and width of the previous layer.

By running sliding windows convolutionaly it becomes much more efficient. You run the convolutions over the full image and the result in the end is equivalent to all results from running separetaly.



### YOLO

YOLO divides the image in SxS regions


### Intersection over Union

It's a way of avaluating object detection by comparing the bounding boxesof label and the one found by the NN. 
 
$$IoU = \frac{size of I}{size of U}$$

"Correct" if IoU >= 0.5


### Non-Max supression

Supresses the overlapping detections. Take down all bounding boxes where $p_c < 0.5$ (or any other chosen probability). Then choose the box with the higher $p_c$ and then take down any detection box with a high IoU with this one.

### Anchor boxes

Sometimes you may have more than 1 object center in a grid cell. To deal with this we use anchor boxes. We put more than one set of $p_c$, $b_x$, ..., $c_3$ in each cells output. Each one of these is associated with an anchor box with a different shape. Now each detected object will outputed in the block of detection with the most similar shaped anchor box.
