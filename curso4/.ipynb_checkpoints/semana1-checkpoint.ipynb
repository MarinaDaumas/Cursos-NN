### Edge detection

    Previous layers detect basic shapes and edges and the later more complex features. 

- Vertical edge exemple:

Convolution between the image and the vertical edge filter



This is filter detects bright-dark edges in grey scale images.


There are ready made commands to do the convolution:

- tensorflow comand: tf.nn.cont2d.



### Padding 

    To avoid getting the image shrinking the image every convolution and using the pixels on the borders less than the one in the middle we add pixels to the borders of the image.
    
    The convention is padding with 0s
    
- **p** is the padding amount (how many pixels are added to each side)
