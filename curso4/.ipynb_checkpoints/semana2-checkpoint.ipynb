### ResNets

    Usefull for very deep NN to avoid vanishing and exploding gradients. (high training error)
    
    They are made up of residual blocks, wich are blocks of layers that have an extra connection that doesn't exist in plain nets
    


the activation of the last layer of a residual block is $a^{[l+2]} = g(z^{[l+2]} + a^{[l]})$

*Why it works?*

In very deep NNs, is commun that a layer worsens the result instead of helping. By using residual blocks it's very easy for the NN to ignore one layer, by making $W*a + b$ of that layer equal to zero and only using the a from a previous layer. Than, $g(a^{[l]}) = a{[l]}$. And there's a chance by gradient descent will make this layer usefull instead of only unharmful.

*What of $a^{[l]}$ and $a^{[l+2]}$ are not tha same size?*

Multiply $a^{[l]}$ with a matrix W_s, so the result is the same dimension as $a^{[l+2]}$. It can be a zero padding matrix or a parameter to be learned.

### Inception Network

    Instead of choosing what layer will a layer be, it tests many different types and shapes and stacks the results of them together.
    
    Down side is it takes a lot of computational power.
    
*How to decrese the number o needed multiplications?*

Add a bottleneck convolution layer. If you have a 28x28x192 and a 5x5x32 filter you can decrease you multiplications by doing a convolution over a 1x1x16 filter and then convolving the result by a 5x5x32 layer and the result will be a same shape layer but doing about 1/10 of the multiplications of the original.

### Transfer learning
 
In the case of computer vision, transfer learning should always be considered, unless you have an extremely large dataset. When doing tranfer learninig, you can download a NN from an open source repository and then, train over it, using the trained weigths of it. Depending on the size of your dataset you may train it diifrently.

- Small dataset

Change only the softmax layer of the original network. Create a new layer with the approriate number of nodes and train only this one. Let all other layers freezed with original weights.

- Not too small dataset

You may change the softmax and a few other previous layers. The previous layers can be used the way they are and just overwrite the weights, or you can completely change these layers.

- Big dataset

Can use just the architecture of the NN and use the original weights as initail parameters but let all layers have their parameters optimized for your purpose.

To make NNs with fixed parts train faster, you can get the output of the last fixed layer for all your training set data and save it, so the computer doesn't have to go through this part every epoch.
